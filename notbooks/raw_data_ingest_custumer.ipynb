```python
total_na = df.isna().sum().sum()
na_variavel = df.isna().sum()

perc_na = na_variavel / total_linhas * 100

df_na = pd.DataFrame({
    'total': total_na,
    'percentual': perc_na
})

print(f"Total de valores Null no DataFrame: {total_na}.\n")
print(df_na)
```

```python
numero_cidades = df["City"].nunique()
cidades = df["City"].unique()

print(f"Número de cidades únicas: {numero_cidades}")
print(f"Cidades únicas: {cidades}")
```

```python
numero_status_civil = df["Marital_Status"].nunique()
status_civil = df["Marital_Status"].unique()

print(f"Número de status civil únicos: {numero_status_civil}")
print(f"Status civil únicos: {status_civil}")
```

```python
numero_target = df["Purchased"].nunique()
target = df["Purchased"].unique()

print(f"Número de target únicos: {numero_target}")
print(f"Target únicos: {target}")
```

```python
inconsistentes = df[(df["Purchased"] < 0) | (df["Purchased"] > 1)]

if not inconsistentes.empty:
    print("Incoerencia de registros:")
    print(inconsistentes)
else:
    print("Dados regulares")
```

Feature engineering

```python
df["Income_per_Age"] = df["Income"] / df["Age"].replace(0, np.nan)
df["Age_x_Score"] = df["Age"] * df["Score"]
```

```python
df = pd.get_dummies(df, columns=["Marital_Status", "City"], drop_first=True)
print(df.head())
```
